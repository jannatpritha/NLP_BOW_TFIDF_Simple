# Preprocessing

```python
# Step 1 Tokenize
tokenized_paragraph = nltk.sent_tokenize(paragraph)
print(tokenized_paragraph)
```

```python
import re
corpus = []
for i in range(len(tokenized_paragraph)):
  review=re.sub('[^a-zA-Z]', ' ',tokenized_paragraph[i]) # remove characters except a-z,A-z
  review= review.lower() # make all lower alphabets
  review=review.split() # make them words
  review=[word for word in review if not word in set(stopwords.words('english'))] # remove stopwords
  review = ' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

#Stemmer

```python
#sample Stemmer
stemmer = PorterStemmer()
stemmer.stem('playing')
```

```python
#stemming
for i in corpus:
  words=  i
  for word in words:
    if word not in set(stopwords.words('english')): #stopwords
      stemmed_paragraph=stemmer.stem(word)
      print(stemmed_paragraph)
```

# Lemmatizer

```python
# sample Lemmatizer
import nltk
nltk.download('wordnet') # Download the WordNet resource

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('playing')
```

```python
#lemmatizing

for i in tokenized_paragraph:
  words=  nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      lemmatized_paragraph=lemmatizer.lemmatize(word)
      print(lemmatized_paragraph)
```

# Complete BOW

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

# Download WordNet data if you haven't already
nltk.download('wordnet')

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()



# Custom tokenizer function
def lemmatize_tokenizer(text):
    words = text.split()
    # Remove stopwords
    words = [word for word in words if word not in stopwords.words('english')]
    return [lemmatizer.lemmatize(word) for word in words]

# Initialize CountVectorizer with custom tokenizer
cv = CountVectorizer(max_features=1500, tokenizer=lemmatize_tokenizer)

# Fit and transform your raw text data (not the tokenized version)
Z = cv.fit_transform(tokenized_paragraph).toarray()
```
